```python
# Reload the original dataset from the source file
df_original = pd.read_csv("spotify_2000.csv")  # Use "spotify_2000.txt" if that's your file

# Group by Artist and count the number of Titles (unfiltered)
artist_counts = df_original.groupby("Artist")["Title"].count().reset_index()
artist_counts.columns = ["Artist", "Song Count"]

# Sort artists by Song Count in descending order
artist_counts = artist_counts.sort_values(by="Song Count", ascending=False)

# Select the top 30 artists
top_artists = artist_counts.head(30)

# Create the horizontal bar chart
plt.figure(figsize=(10, 12))  # Set figure size for 30 artists
plt.barh(top_artists["Artist"], top_artists["Song Count"], color='lightblue')  # Solid blue bars
plt.xlabel("Count of Title")
plt.ylabel("Artist")
plt.title("Count of Title by Artist")

# Set x-axis ticks dynamically
max_count = top_artists["Song Count"].max()
plt.xlim(0, max_count + 5)
plt.xticks(np.arange(0, max_count + 10, 10))

# Add light gray gridlines on the x-axis
plt.grid(axis='x', color='lightgray')

# Display the plot
plt.show()
```

```python
# Compare loudness and danceability for popular songs
popular_threshold = 75  # Define popularity threshold for "popular"
popular_songs = df[df['Popularity'] >= popular_threshold]
# Calculate average loudness and danceability for popular songs
avg_loudness = popular_songs['Loudness (dB)'].mean()
avg_danceability = popular_songs['Danceability'].mean()
# Create a DataFrame with the result
df = pd.DataFrame({'Answer': ['Loud' if avg_loudness > avg_danceability else 'Danceable']})
display(df)
```
